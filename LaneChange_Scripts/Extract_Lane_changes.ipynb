## Constants

```python
LANGECHANGE_AND_BREAKING_DATA_PATH = "data/braking_and_lanechange.csv"
BRAKING_CUT = "data/braking_cut.csv"
EIGHT_BRAKING_CUTS = "data/braking_cut_8_brakes.csv"
BRAKING_ONE_ROUND = "data/braking_eine_Runde.csv"
LC_ONE_ROUND = "data/LC_eine_Runde.csv"
```

```python
LCandBrakingDF = pd.read_csv(LANGECHANGE_AND_BREAKING_DATA_PATH, sep=";")
brakingCut = pd.read_csv(BRAKING_CUT, sep=";")
eightBrakingCuts = pd.read_csv(EIGHT_BRAKING_CUTS, sep=";")
brakingOneRound = pd.read_csv(BRAKING_ONE_ROUND, sep=";")
lcOneRound = pd.read_csv(LC_ONE_ROUND, sep=";")

dataframes = [LCandBrakingDF, brakingCut, eightBrakingCuts, brakingOneRound, lcOneRound] # put into an array for simplifying data processing
```
