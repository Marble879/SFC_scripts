## Constants

## Initialize data frame

```python
LCandBrakingDF = pd.read_csv(LANGECHANGE_AND_BREAKING_DATA_PATH, sep=";", index_col=False)
brakingCut = pd.read_csv(BRAKING_CUT, sep=";", index_col=False)
eightBrakingCuts = pd.read_csv(EIGHT_BRAKING_CUTS, sep=";", index_col=False)
brakingOneRound = pd.read_csv(BRAKING_ONE_ROUND, sep=";", index_col=False)
lcOneRound = pd.read_csv(LC_ONE_ROUND, sep=";", index_col=False)

dataframes = [LCandBrakingDF, brakingCut, eightBrakingCuts, brakingOneRound, lcOneRound] # put into an array for simplifying data processing
```

### verify 8 brakes is a subset of lanechange and brakes

```python
sortedEightBrakingCuts = eightBrakingCuts.sort_values(by=list(eightBrakingCuts.columns)).reset_index(drop=True)
sortedLCandBrakingDF = LCandBrakingDF.sort_values(by=list(LCandBrakingDF.columns)).reset_index(drop=True)

set1 = set(map(tuple, sortedEightBrakingCuts.to_numpy()))
set2 = set(map(tuple, sortedLCandBrakingDF.to_numpy()))
print(set1.issubset(set2))
```

```python
### Remove the 8 brakes csv from the lane change & brakes csv
```

```python
merged_df = LCandBrakingDF.merge(eightBrakingCuts, indicator=True, how='inner')
matched_rows = merged_df[merged_df['_merge'] == 'both']
lcDF = LCandBrakingDF.drop(matched_rows.index).reset_index(drop=True)
lcDF
lcDF.to_csv("data/lane_changes.csv", index=False)
```

```python
# confirming if brake data is in lcdf
sortedEightBrakingCuts = eightBrakingCuts.sort_values(by=list(eightBrakingCuts.columns)).reset_index(drop=True)
sortedlcDF = lcDF.sort_values(by=list(lcDF.columns)).reset_index(drop=True)

set1 = set(map(tuple, sortedEightBrakingCuts.to_numpy()))
set2 = set(map(tuple, sortedlcDF.to_numpy()))
print(set1.issubset(set2))
```
